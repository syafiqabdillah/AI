```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import os

CATEGORIES = ["Cat", "Dog"]

# method to prepare the data before prediction 
def prepare(filepath):
    try:
        IMG_SIZE = 50 # resize the image 
        img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) # grayscale the image
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
    except Exception as e:
        pass
```

```python
# the images we want to predict 

DATADIR = "D:\Artificial Intelligence\Inta_img"

```

```python
# loading the model 
model = tf.keras.models.load_model("catdog.model") # loading the model
```

```python
# make prediction for each image, while displaying the image 
for img in os.listdir(DATADIR):  
    try:
        prediction = model.predict([prepare(os.path.join(DATADIR, img))]) # the prediction
        print(CATEGORIES[int(prediction[0][0])])
        img_array = cv2.imread(os.path.join(DATADIR, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray")
        plt.show() # the actual image 
    except Exception as e:
        pass
```
