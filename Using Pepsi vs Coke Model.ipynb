```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import os

CATEGORIES = ["Pepsi", "Coke"]

# method to prepare the data before prediction 
def prepare(filepath):
    try:
        IMG_SIZE = 70 # resize the image 
        img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) # grayscale the image
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
    except Exception as e:
        pass
```

```python
# the images we want to predict 

DATADIR = "D:\Artificial Intelligence\Pepsi Coke Model Test"

```

```python
# loading the model 
model = tf.keras.models.load_model("pepsicoke.model") # loading the model
```

```python
# make prediction for each image, while displaying the image 
for img in os.listdir(DATADIR):  
    try:
        prediction = model.predict([prepare(os.path.join(DATADIR, img))]) # the prediction
        print(CATEGORIES[int(prediction[0][0])])
        # show the prepared image 
        
        img_array = cv2.imread(os.path.join(DATADIR, img), cv2.IMREAD_GRAYSCALE) 
        new_array = cv2.resize(img_array, (70, 70))
        
        plt.imshow(new_array, cmap="gray")
        plt.show() 
    except Exception as e:
        print(e)
        pass
```
